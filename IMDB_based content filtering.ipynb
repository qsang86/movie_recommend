```python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')

movies = pd.read_csv('./data/tmdb_5000_movies.csv')
credits = pd.read_csv('./data/tmdb_5000_credits.csv')
print(movies.shape)
movies.head(5)
```

```python
from ast import literal_eval

movies_df = movies[['id', 'title', 'genres', 'vote_average', 'vote_count', 'popularity', 'keywords', 'overview']]
movies_df.head(5)
```

```python
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)

movies_df['genres'] = movies_df['genres'].apply(lambda x : [ y['name'] for y in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x : [ y['name'] for y in x])

```

```python
movies_df[['genres', 'keywords']][:1]
##movies_df['genres'].head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
count_vec = CountVectorizer(min_df=0, ngram_range=(1, 2))
genre_mat = count_vec.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

```python

genre_similarity = cosine_similarity(genre_mat, genre_mat)
print(genre_similarity.shape)
print(genre_similarity[:1])

genre_similarity_sorted = genre_similarity.argsort()[:, ::-1]
print(genre_similarity_sorted[:1])
```

## Genre-based content filtering

```python
def find_sim_movies(df, sorted_index, title_name, top_n = 10):
    title_moive = df[df['title']== title_name]
    
    title_index = title_moive.index.values
    sim_indexes = sorted_index[title_index, :(top_n)]
    
    print(sim_indexes)
    sim_indexes = sim_indexes.reshape(-1)
    
    return df.iloc[sim_indexes]
```

## weighted rating used by IMDB

WR = (v/(v+m))*r + (m/(v+m))*c

r = mean voting rate for individual movies

v = number of votes for individual movies

m = minimum votes required to be listed in the Top Rated List

c = mean vote acrosee the whole report (all movies)


referred from "https://help.imdb.com/article/imdb/track-movies-tv/ratings-faq/G67Y87TFYYP6TWAV#"


```python
c = movies_df['vote_average'].mean()

# set as the top 70% count - because I love number 7
percentile = 0.7
m = movies_df['vote_count'].quantile(percentile)

print("c: ", c)
print("m: ", m)
```

```python
c = movies['vote_average'].mean()

percentile = 0.7
m = movies['vote_count'].quantile(percentile)

def weighted_rating(input_data):
    r = input_data['vote_average']
    v = input_data['vote_count']
    
    weighted_rating_result = (v/(v+m))*r + (m/(v+m))*c
    
    return weighted_rating_result

movies['weighted_vote'] = movies.apply(weighted_rating, axis=1)

movies_df['weighted_vote'] = movies_df.apply(weighted_rating, axis=1)
movies_df[['title', 'genres', 'vote_average', 'weighted_vote', 'vote_count']].sort_values('weighted_vote', ascending=False).head(15)
```

```python
def find_sim_movies_wr(df, sorted_index, title_name, top_n = 10):
    title_moive = df[df['title']== title_name]
    
    title_index = title_moive.index.values
    sim_indexes = sorted_index[title_index, :(top_n)]
    
    print(sim_indexes)
    sim_indexes = sim_indexes.reshape(-1)
    
    return df.iloc[sim_indexes].sort_values('weighted_vote', ascending=False).head(top_n)


```

```python
from random import randint

value = randint(0, len(movies_df))
#print(value)
rand_title = movies_df.loc[value]['title']

final_recommend = find_sim_movies_wr(movies_df, genre_similarity_sorted, rand_title, 10)

print('\033[1m' + "Looking for movies similar to this?: ", rand_title)
final_recommend[['id','title', 'genres', 'vote_average', 'weighted_vote', 'vote_count']].head(5)
```
